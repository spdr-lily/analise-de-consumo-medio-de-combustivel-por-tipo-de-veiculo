```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Dados de consumo de combustível:
dados_combustivel = {
    'tipo_veiculo':['SUV', 'Compacto', 'Sedan', 'Hatchback', 'SUV', 'Compacto'],
    'consumo':[9.5, 12.3, 10.1, 11.8, 8.7, 13.2]
}
```

```python
# Criar o DataFrame a pártir dos dados
df_combustivel = pd.DataFrame(dados_combustivel)
```

```python
# visualizar os dados
print(df_combustivel)
```

```python
# calcular a média de consumo por tipo de veículo
media_consumo = df_combustivel.groupby('tipo_veiculo')['consumo'].mean()
```

```python
# plotar um gráfico de barras com a média de consumo por tipo de veículo
plt.bar(media_consumo.index, media_consumo)
plt.xlabel('Tipo de Veículo')
plt.ylabel('Consumo Médio (km/1)')
plt.title('Consumo Médio de Combustível por Tipo de Veículo')
plt.show()
```

1. Os dados de consumo de combustível são armazenados em um DataFrame do Pandas.<br>
2. Os dados são inseridos diretamente no código, em um dicionário chamado dados_combustivel. Cada entrada do dicionário contém as informações de 'tipo_veiculo' e 'consumo'.<br>
3. O código calcula a média de consumo de combustível por tipo de veículo utilizando a função groupby do Pandas.<br>
4. O gráfico de barras exibe a média de consumo de combustível por tipo de veículo. O eixo x representa os tipos de veículos, e o eixo y representa o consumo médio em quilômetros por litro.<br>
5. O código utiliza as bibliotecas Pandas para a análise de dados e Matplotlib para a vizualização do gráfico de barras.
